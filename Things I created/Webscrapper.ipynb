# Magalu

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.88 Safari/537.36'}
response2 = requests.get('https://www.magazineluiza.com.br/pacote-office/informatica/s/in/paof/', headers=headers)
soup2 = BeautifulSoup(response2.content, "html.parser")
title2 = []
link2 = []
price2 = []
today = date.today()
url = []
urls = []
#Getting the name and the price
price = []
title = []

for i in soup2.find_all('a', href=True):
     url.append(i['href'])
for i in url:
    if i.endswith('/in/paof/'):
        if i != '/pacote-office/informatica/s/in/paof/':
            urls.append(i)

for url in urls:
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")

    for i in soup.find_all('h1', class_="header-product__title"):
        title.append(i.get_text())
    for i in soup.find_all('span',class_="price-template__text"):
        price.append(i.get_text())

#Making it pretty

today = date.today()
df1 = pd.DataFrame(title)
df1['Preço'] = pd.DataFrame(price)
df1['Urls'] = pd.DataFrame(urls)
df1['Data'] = today.strftime("%m/%d/%Y")
df1['Varejo'] = 'Magalu'
df1.columns = ['Nome', 'Preço', 'Url', 'Data', 'Varejo']
```
